```python
import cv2
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
blank_img = np.zeros(shape=(512,512,3), dtype=np.int16)
```

```python
blank_img.shape
```

```python
plt.imshow(blank_img)
```

```python
# lets draw a rectangle on the image 
cv2.rectangle(blank_img, pt1=(100,30),pt2=(400,400),color=(0,255,0),thickness=10)
plt.imshow(blank_img)
```

```python
cv2.rectangle(blank_img, pt1=(200,200),pt2=(300,300),color=(0,0,255),thickness=10)
plt.imshow(blank_img)
```

```python
cv2.circle(blank_img, center=(340,300),radius=50,color=(255,0,0),thickness=-1)
plt.imshow(blank_img)
```

```python
cv2.line(blank_img,pt1=(0,0),pt2=(512,512),color=(255,255,255),thickness=5)
plt.imshow(blank_img)
```

```python
# NOW LETS USE A FONT TO WRITE ON THE IMAGE
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(blank_img,text='Hello',org=(10,500),fontFace=font,fontScale=4,color=(255,255,255),thickness=3,lineType=cv2.LINE_AA)
plt.imshow(blank_img)
```

```python
# NOW LETS CREATE A POLYGON
new_img = np.zeros(shape=(512,512,3), dtype=np.int16)
```

```python
vertices = np.array([[100,100],[200,200],[400,350],[200,400]],dtype=np.int32)
vertices.shape
```

```python
pts=vertices.reshape((-1,1,2))
```

```python
pts.shape
```

```python
cv2.polylines(new_img,[pts],isClosed=True,color=(0,255,0),thickness=10)
plt.imshow(new_img)
```
